```python
## create a simple text file

import os
os.makedirs("../data/text_files",exist_ok=True)
```

```python
sample_texts={
    "../data/text_files/python_intro.txt":"""python programming Introduction
    Python is a high-level, general-purpose, and interpreter-based programming language widely used for software development, 
    data science, automation, artificial intelligence, and web development.It was created by Guido van Rossum and first released 
    in 1991 with the goal of providing a language that emphasizes code readability and developer productivity.One of Python’s 
    strongest features is its simple and clean syntax, which resembles natural language and allows beginners to learn programming easily. 

    Key Features of Python

        1.Simple and Easy Syntax- Python’s syntax is clean and similar to English, making it easy to learn and read.
        2.High-Level Language - Developers can write programs without worrying about low-level hardware details.
        3.Object-Oriented Programming -Supports classes, objects, inheritance, polymorphism, and encapsulation.
        4.Large Standard Library-Comes with built-in modules for file handling, math operations, networking, etc.
        5.Platform Independent- Python programs can run on Windows, macOS, Linux without changes.
        6.Dynamic Typing - No need to declare variable types; Python decides the type during runtime.
        7.Automatic Memory Management - Garbage collection is built in, so memory is handled automatically.
        8.Extensible and Embeddable - Python can be combined with C, C++, Java for performance improvements.
        9.Supports Multiple Paradigms-Procedural, Functional, and Object-Oriented styles.
        10.Interpreted Language- Python executes code line by line, making debugging easier.

    Python is platform-independent, meaning programs written in Python can run on Windows, macOS, and Linux without modification. 
    Additionally, Python’s dynamic typing, automatic memory management, and extensive community support make it one of the most popular
    and powerful programming languages in the world today."""
}    

for filepath,content in sample_texts.items():
    with open(filepath,"w",encoding="utf-8") as f:
        f.write(content)

print("Sample text files created successfully.")
```

```python
### TextLoader -->read




from langchain_community.document_loaders import TextLoader
```

```python
loader=TextLoader("../data/text_files/python_intro.txt",encoding="utf-8")
document=loader.load()
print(document)
```

```python
##directory loader
from langchain_community.document_loaders import DirectoryLoader

#load all text files from a directory
dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",##pattern to match files
    loader_cls=TextLoader,#loader class to use
    loader_kwargs={"encoding":"utf-8"},
    show_progress=False    
    
)

documents=dir_loader.load()
print(documents)


for doc in documents:
    print(doc.page_content)
    print("="*80)
```

```python
from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader
dir_loader =DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
    
    show_progress=False
)

pdf_documents=dir_loader.load()
print(pdf_documents)
```
